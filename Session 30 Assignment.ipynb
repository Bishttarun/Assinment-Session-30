```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from sklearn import cluster, datasets
import scipy.misc
face = scipy.misc.face(gray=False)
```

```python
face.shape
```

```python
plt.figure(figsize=(8,8), dpi=100)
plt.imshow(face)
```

```python
face_reshape = (face / 255).reshape(-1,3)
```

```python
face_reshape.shape
```

```python
from sklearn.cluster import KMeans
k_colors = KMeans(n_clusters=5).fit(face_reshape)
y_pred=k_colors.predict(face_reshape)
```

```python
k_colors.cluster_centers_.shape
```

```python
k_colors.labels_.shape
```

```python
newimg=k_colors.cluster_centers_[k_colors.labels_]
```

```python
newimg.shape
```

```python
newimg=np.reshape(newimg, (face.shape))
```

```python
newimg.shape
```

```python
fig = plt.figure(figsize=(20,20))
ax=fig.add_subplot(1,2,1,xticks=[],yticks=[],title='Original Image')
ax.imshow(face)
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title='Color Compressed Image using K-Means with 5 clusters')
ax.imshow(newimg)
plt.show()
```
